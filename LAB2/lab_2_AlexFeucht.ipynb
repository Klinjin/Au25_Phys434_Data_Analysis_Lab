````python
# TODO: Step 1 - Generate 1000 uniform random samples between 0 and 1
# Use scipy.stats.uniform or numpy.random.uniform

# TODO: Step 2 - Apply transformation u = x^2

# TODO: Step 3 - Plot histogram of transformed values
# Also plot the analytical solution g(u)

# TODO: Step 4 - Add text explaining the analytical derivation
# For u = x^2, with x ~ Uniform(0,1), derive g(u)
````

````python
# Uniform distribution of 1000 samples between 0 and 1
uniformSample = np.random.uniform(0, 1, 1000)

# Transformed uniform distribution [u = x^2]
transformU = np.power(uniformSample, 2)

# Analytic distribution [g(u) for u = x^2, x ~ Uniform(0,1) is g(u) = 1/(2*sqrt(u)) for 0 < u < 1]
u_values = np.linspace(0.01, 1, 100) # Avoid u=0 for sqrt
analyticDist = 1/(2*np.sqrt(u_values))

# Create figure histograms
fig, ax = plt.subplots(figsize=(10, 6))
ax.hist(uniformSample, bins = 40, density=True, label=f'Uniform')
ax.hist(transformU, bins = 40, density=True, alpha=0.6, label=f'Transformed')

# Plot the analytic solution on top of histograms
ax.plot(u_values, analyticDist, 'r-', lw=2, label='Analytic g(u)')

# Set labels and title
ax.set_title(f"Uniform vs. Transformed Histograms and Analytic Solution")
ax.set_xlabel("Value")
ax.set_ylabel("Density")
ax.legend()
plt.show()
````

````python
# TODO: Step 1 - Simulate throwing two dice N=1000 times
# Use numpy.random.randint(1, 7, size=(1000, 2)) or equivalent method

# TODO: Step 2 - Calculate sum of each pair

# TODO: Step 3 - Plot distribution of sums

# TODO: Step 4 - Calculate mean and variance

# TODO: Step 5 - Calculate probability of sevens and statistical uncertainty

# TODO: Step 6 - Compare with analytical calculation (1/6 probability) on a histogram plot
````

````python
# Create a dice roll sample of 2d6 rolled 1000 times
dice1 = np.random.randint(1, 7, size=(1000, 2))
# print(len(dice1))

# Sum values of all dice in dice1 array along the 1 axis
sum1 = np.sum(dice1, axis=1)
# print(len(sum1))
# print(sum1[:100])

# Create figure
fig, ax = plt.subplots(figsize=(10, 6))

# Plot histogram with density=True to compare with probability
ax.hist(sum1, bins=np.arange(2, 14) - 0.5, density=True, label=f'Sum of Pairs (Simulation)')
ax.set_title("1000 Rolls of 2 6-Sided Dice")
ax.set_xlabel("Dice Roll Value")
ax.set_ylabel("Probability Density")

# Analytic Calculation for Probability of 7
analytic_prob_7 = 1/6
ax.axhline(analytic_prob_7, color='r', linestyle='dashed', linewidth=2, label=f'Analytic Probability of 7 ({analytic_prob_7:.4f})')
ax.legend()
plt.show()

# Mean
mu1 = sum(sum1)/len(sum1)
print(f"Mean of sums: {mu1}")

# Variance
var1 = sum((sum1 - mu1)**2)/len(sum1)
print(f"Variance of sums: {var1}")

# Probability of 7s
simulated_prob_7 = sum(sum1 == 7)/len(sum1)
print(f"Simulated probability of 7s: {simulated_prob_7}")

# Statistical Uncertainty of the probability (assuming Binomial errors)
# For Binomial distribution, variance is n*p*(1-p)
# Standard error of the proportion is sqrt(p*(1-p)/n)
n_rolls = len(sum1)
p_analytic = 1/6
statistical_uncertainty = np.sqrt(p_analytic * (1 - p_analytic) / n_rolls)
print(f"Statistical uncertainty of simulated probability of 7s (Binomial error): {statistical_uncertainty:.4f}")

# Compare simulation with analytical calculation
print(f"Difference between simulated and analytic probability of 7s: {simulated_prob_7 - analytic_prob_7:.4f}")
````

# Central Limit Theorem (4 points)

The central limit theorem states that the distribution of the sample mean $\bar{x}$,
$\bar{x} = \frac{1}{N} \sum_{i=1}^{N} x_i$,
of N variables $x_i$ with finite variance $\sigma^2$ in the limit $N\to \infty$ will approach a normal distribution with variance $\sigma^2/N$ independent of the form of the distribution f(x). In this computational experiment, we will test this theorem.

Generate a M=5 event sample using built-in Poisson distributions in scipy with $\lambda=2.0$. From these 5 events, calculate mean $\mu_5$ and standard deviation $\sigma_5$. Repeat this process N times. For each time, you will draw different set of M=5 events, and calculate independent $\mu_5$ and $\sigma_5$, respectively.

## Part 1: N=10 samples

````python
# TODO: Let N=10. Generate N sets of M=5 Poisson samples with lambda=2.0
# Calculate mu_5 and sigma_5 for each set
# Plot distributions of mu_5 and sigma_5
# Include mean and variance in the legend
````

````python
N = 10 # repeat sampling
M = 5 # event number
lambda_val = 2.0 # effective width of distribution about mean

# Generate N sets of M=5 Poisson samples with lambda=2.0
poisson_samples = np.random.poisson(lambda_val, size=(N, M))

# Calculate mu_5 and sigma_5
mu_5 = np.mean(poisson_samples, axis=1)
sigma_5 = np.std(poisson_samples, axis=1, ddof=1)
mean = np.mean(poisson_samples)
var = np.var(poisson_samples)

# Plot distributions of mu_5 and sigma_5
fig, ax = plt.subplots(figsize=(10, 6))

# Plot histogram of mu_5 (means of the sets of 5)
ax.hist(mu_5, bins=10, density=True, label=(f"Distribution of mu_5 for N={N}, M={M}"))

# Plot histogram of sigma_5 (standard deviations of the sets of 5)
ax.hist(sigma_5, bins=10, density=True, alpha=0.6, label=(f"Distribution of sigma_5 for N={N}, M={M}"))

# Set labels and title
ax.set_title(f"Distribution of Sample Means (mu_5) for Poisson Distribution (lambda={lambda_val}), N={N} Samples of M={M}")
ax.set_xlabel("Sample Mean (mu_5)")
ax.set_ylabel("Density")
ax.legend()
plt.show()

# Calculate and print mean and variance of mu_5 and sigma_5 distributions
mean_mu_5 = np.mean(mu_5)
var_mu_5 = np.var(mu_5)
mean_sigma_5 = np.mean(sigma_5)
var_sigma_5 = np.var(sigma_5)

print(f"Mean of Poisson samples: {mean}")
print(f"Variance of Poisson samples: {var}")
print(f"Mean of mu_5 distribution: {mean_mu_5}")
print(f"Variance of mu_5 distribution: {var_mu_5}")
print(f"Mean of sigma_5 distribution: {mean_sigma_5}")
print(f"Variance of sigma_5 distribution: {var_sigma_5}")
````
